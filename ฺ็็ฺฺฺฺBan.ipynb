```python
%matplotlib inline
import matplotlib.pyplot as plt
import seaborn as sns; sns.set()
import numpy as np
import pandas as pd
rng = np.random.RandomState(1)
x = 10 * rng.rand(30)
y = 2 * x - 5 + rng.randn(30)
#y = -2 * x + 15 + rng.randn(30)
plt.scatter(x, y);
```

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression()
Xtrain = x[:, np.newaxis]
model.fit(Xtrain, y)
LinearRegression(copy_X=True, fit_intercept=True, n_jobs=None)
a = model.coef_[0]
a
b = model.intercept_
b
inputx = float(input('ค่า x = '))
Ynew = a*inputx + b
print('you should have ynew = ', Ynew)
Ynew = a*x + b
plt.scatter(x, y)
plt.plot(x, Ynew)
```

```python
df = pd.read_csv("baan4000.csv")
df.head()
pro = input('จังหวัด? : ')
data = df[ df.province == pro]
data = data[ data.area < 1000 ]
print(data.size)
area = data[ ['area'] ]
price = data[ ['price'] ]
x = area.values
y = price.values.ravel()
model = LinearRegression()
model.fit(x, y)
a = model.coef_[0]
b = model.intercept_
Ynew = a*x + b
plt.scatter(area, price)
plt.plot(x, Ynew)
พื้นที่ = float(input('บ้านมีที่ดินกี่ตารางวา? : '))
Ynew = a*พื้นที่ + b
print(f"คุณควรขายที่ดินราคา {Ynew:,.2f} บาท")
```
